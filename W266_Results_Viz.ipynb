```python
#Visualization of cross training results
import matplotlib.pyplot as plt
```

```python
#Cross training data.  Listed as " trained on - evaluated on"
CT1_exact = [56.1, 11.57, .10, 72.83, 59.44]
CT1_f1 = [61.39, 14.64, .23, 76.23, 62.55]
CT1_names = ['NQ - NQ', 'NQ - Trivia QA', 'NQ, QA - QA', 'NQ,QA - SQuAD', 'NQ,QA,SQuAD - SQuAD']
```

```python
plt.figure(figsize=(8, 5))

plt.plot(CT1_names, CT1_exact, 'b.--', label = 'EM')
plt.plot(CT1_names, CT1_f1, 'g.-', label = 'F1')
plt.xlabel('Dataset Trained - Dataset Evaluated')
plt.xticks(rotation = 45)
plt.ylabel('Metric Score')
plt.title('Cross Training Experiment One - DistilBert QA')
plt.legend(loc = 'lower right')
```
